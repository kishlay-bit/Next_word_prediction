```python
quotes = quotes.str.lower()
```

```python
import string
translator = str.maketrans('', '', string.punctuation)
quotes = quotes.apply(lambda x: x.translate(translator))

quotes.head()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

vocab_size = 8978

tokinizer = Tokenizer(num_words=vocab_size)
tokinizer.fit_on_texts(quotes)
```

```python
word_index = tokinizer.word_index
print(len(word_index))
list(word_index.items())[:10]
```

```python
sequence = tokinizer.texts_to_sequences(quotes)
```

```python
for i in range(3):
  print(quotes[i])
```

```python
for i in range(3):
  print(sequence[i])
```

```python
X = []
y = []

for seq in sequence:
  for i in range(1,len(seq)):
    input_seq = seq[:i]
    output_seq = seq[i]
    X.append(input_seq)
    y.append(output_seq)
```

```python
X
```

```python
len(X)
```

```python
len(y)
```

```python
max_len=max(len(x) for x in X)
print(max_len)
```

```python

from tensorflow.keras.preprocessing.sequence import pad_sequences
X_padded = pad_sequences(X, maxlen=max_len, padding='pre')
```

```python
X_padded
```

```python
y=np.array(y)
```

```python
from tensorflow.keras.utils import to_categorical
y_one_hot = to_categorical(y, num_classes=vocab_size)
```

```python
y.shape
```

```python
y_one_hot.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,LSTM, Dense
```

```python
embedding_dim = 50
rnn_units = 128
```

```python
rnn_model=Sequential()
rnn_model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len))
rnn_model.add(
    SimpleRNN(units=rnn_units, activation='tanh', return_sequences=False)
)
rnn_model.add(Dense(units=vocab_size, activation='softmax'))
```

```python
rnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
rnn_model.summary()
```

```python
lstm_model = Sequential()
lstm_model.add(
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len)
)
lstm_model.add(LSTM(units=rnn_units))
lstm_model.add(Dense(units=vocab_size, activation='softmax'))
```

```python
lstm_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
lstm_model.summary()
```

```python
epochs=100
batch_size=128
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# 1. Define the early stopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',     # Quantity to be monitored
    patience=5,             # Number of epochs with no improvement after which training will be stopped
    restore_best_weights=True # Restores model weights from the epoch with the best value
)
```

```python
history_rnn = rnn_model.fit(
    X_padded,
    y_one_hot,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2,
    callbacks=[early_stopping]
)
```

```python
history_lstm = lstm_model.fit(
    X_padded,
    y_one_hot,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2,
    callbacks=[early_stopping]
)
```

```python
index_to_word = {}
for word, index in word_index.items():
  index_to_word[index] = word
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def predictor(model,tokenizer,text,max_len):
  text = text.lower()

  seq = tokenizer.texts_to_sequences([text])[0]
  seq = pad_sequences([seq], maxlen=max_len, padding='pre')

  pred = model.predict(seq,verbose = 0)
  pred_index = np.argmax(pred)
  return index_to_word[pred_index]
```

```python

seed_text = "what are you"
next_word = predictor(lstm_model,tokinizer,seed_text,max_len)
print(next_word)
```

```python
def generate_text(model,tokenizer,seed_text,max_len,n_words):
  for _ in range(n_words):
    next_word = predictor(model,tokenizer,seed_text,max_len)
    if next_word == "":
      break
    seed_text += " " + next_word
  return seed_text
```

```python
seed = "are you a "
generate_text = generate_text(lstm_model,tokinizer,seed,max_len,10)
print(generate_text)
```
